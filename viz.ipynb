```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data
data = pd.read_csv('property.csv')


# x = data['Location']
# y = data['Price']
# # print(x, y)
# # Create a figure and axis
# fig, ax = plt.subplots()

# # use LARGE scale so it makes sense
# ax.scatter(x, y, s=100)

# # Plot the data
# plt.plot(fig, ax)


# clean data
def convert_area(area):
    if 'Marla' in area:
        return float(area.replace('Marla', '')) * 272.25
    elif 'Kanal' in area:
        return float(area.replace('Kanal', '')) * 5445
    else:
        return float(area)
    

data['Area'] = data['Area'].apply(convert_area)

def convert_price(price):
    if 'Crore' in price:
        return float(price.replace('Crore', '')) * 10000000
    elif 'Lakh' in price:
        return float(price.replace('Lakh', '')) * 100000
    elif 'Arab' in price:
        return float(price.replace('Arab', '')) * 1000000000
    else:
        return float(price)
    
data['Price'] = data['Price'].apply(convert_price)

# remove outliers
data = data[data['Area'] < 10000]
data = data[data['Price'] < 250000000]
```

```python
x = data['Area']
y = data['Price']

# plt.scatter(x, y)
# plt.plot(x, y)
# plt.bar(x, y)
# plt.legend()
```

```python
# add labels
plt.xlabel('Area (sqft)')
plt.ylabel('Price (PKR 1000000)')
plt.title('Price vs Area')
plt.scatter(x, y)
# plt.plot(x, y)
```
